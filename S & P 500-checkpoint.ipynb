### Other co-relative analysis(odf)
* analyse whether Daliy change in Closing price of stocks (or) Daliy Returns in stock are co-related or not !

```python
odf = pd.DataFrame()
```

```python
odf['AAPL_close_pct_change'] =(df['apple_close'] - df['apple_close'].shift(1)) / df['apple_close'].shift(1) * 100
odf['AMZN_close_pct_change'] =(df['amazon_close'] - df['amazon_close'].shift(1)) / df['amazon_close'].shift(1) * 100
odf['GOOG_close_pct_change'] =(df['google_close'] - df['google_close'].shift(1)) / df['google_close'].shift(1) * 100
odf['MSFT_close_pct_change'] =(df['microsoft_close'] - df['microsoft_close'].shift(1)) / df['microsoft_close'].shift(1) * 100
```

```python
odf
```

```python
g = sns.PairGrid(data=odf)
g.map_diag(sns.histplot)
g.map_lower(sns.scatterplot)
g.map_upper(sns.kdeplot) # kernal Density Estimation
```
